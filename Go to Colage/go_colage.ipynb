## possibility of students to university use tensorflow

```python
import pandas as pd
```

```python
import zipfile
local_file = 'archive (1).zip'
zip_ref = zipfile.ZipFile(local_file, 'r')
zip_ref.extractall('data')
zip_ref.close()
```

```python
df = pd.read_csv('data/data.csv')
df.sample(10)
```

```python
df.dtypes
```

```python
def print_unique_col_value(df):
    for column in df:
        print(f'{column}:{df[column].unique()}')
```

```python
print_unique_col_value(df)
```

```python
df.type_school.replace({'Academic':1,'Vocational':0},inplace=True)
df.school_accreditation.replace({'A':1,'B':0},inplace=True)
df.residence.replace({'Urban':1,'Rural':0},inplace=True)
df['parent_was_in_college'].replace({'True':1,'False':0},inplace=True)
df['gender'].replace({'Female':0,'Male':1},inplace=True)
```

```python
print_unique_col_value(df)
```

```python
df2 = pd.get_dummies(data=df,columns=['interest'])
df2.sample(5)
```

```python
df2.dtypes
```

```python

df2.replace({False: 0, True: 1}, inplace=True)
for column in df2:
    print(f'{column}:{df2[column].unique()}')
```

```python
df2.dtypes
```

```python
df2.sample(5)
```

```python
cols_to_scale = ['parent_age','parent_salary','house_area','average_grades']
from sklearn.preprocessing import MinMaxScaler 
scaler = MinMaxScaler()

df2[cols_to_scale] = scaler.fit_transform(df2[cols_to_scale])
```

```python
for column in df2:
    print(f'{column}:{df2[column].unique()}')
```

```python
df2.sample(5)
```

```python
x  = df2.drop("residence",axis = 'columns')
y = df2["residence"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=5)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
"""
import numpy as np
x_train = np.asarray(x_train).astype('float32').reshape((-1,1))
y_train = np.asarray(y_train).astype('float32').reshape((-1,1))
y_train.shape
"""
```

```python
x_train.shape
```

```python
len(x_train.columns)
```

```python
import tensorflow as tf
from tensorflow import keras


model = keras.Sequential([
    keras.layers.Dense(26, input_shape=(14,), activation='selu'),
    keras.layers.Dense(128, activation="selu"),
    keras.layers.Dense(64, activation="selu"),
    keras.layers.Dense(1)  
])

opt = keras.optimizers.Adam(learning_rate=0.0001)

model.compile(optimizer= opt,
              loss='mean_squared_error',
              metrics=['mae'])

model.fit(x_train, y_train, epochs=100)
```

```python
yp = model.predict(x_test)
yp[:5]
```

```python
y_test[:5]
```

```python
ypred = []
for el in yp:
    if el > 0.5:
        ypred.append(1)
    else:
        ypred.append(0)

ypred[:5]
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(classification_report(y_test,ypred))
```

```python
import seaborn as sn
from matplotlib import pyplot as plt
cm = tf.math.confusion_matrix(labels=y_test,predictions=ypred)

plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Urban')
plt.ylabel('Rural')
```
